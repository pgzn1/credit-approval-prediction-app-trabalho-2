# Passo 2. Recuperação dos dados

```python
def download_data(data_url, data_path, data_file):
    os.makedirs(data_path, exist_ok = True)
    local_path = os.path.join(data_path, data_file)
    urllib.request.urlretrieve(data_url, local_path)
```

```python
download_data(CRX_URL, DATA_PATH, DATA_FILE)
```

```python
def load_data(data_path, data_file):
    local_path = os.path.join(data_path, data_file)
    return pd.read_csv(local_path, header=None, names=['Male','Age','Debt','Married','BankCustomer','EducationLevel',
                                                      'Ethnicity','YearsEmployed','PriorDefault','Employed',
                                                      'CreditScore','DriversLicense','Citizen','ZipCode',
                                                      'Income','Approved'])
```

# Passo 4. Tratamento dos Dados

```python
for col in crx_data.columns:
    # Checa se a coluna é do tipo object
    if crx_data[col].dtypes == 'object':
        # Trocar o valor nulo para o valor mais frequente
        crx_data = crx_data.fillna(crx_data[col].value_counts().index[0])
```

## Pré Processamento dos dados

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

for col in crx_data.columns:
    # Compara se o tipo da coluna é object
    if crx_data[col].dtypes=='object':
    # Usa o LabelEnconder para transformar em uma coluna numérica
        crx_data[col]=le.fit_transform(crx_data[col])
```

# Passo 5. Transformação dos dados

```python
from sklearn.preprocessing import MinMaxScaler
#from sklearn.preprocessing import StandardScaler

X = crx_data.drop('Approved', axis=1)
y = crx_data['Approved']

scaler = MinMaxScaler() 
X_scaled = pd.DataFrame(scaler.fit_transform(X))
X_scaled.columns = X.columns
```

# Passo 5. Treinamento do Modelo

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=.3,stratify=y,random_state=RANDOM_STATE)
```

```python
def build_classifiers():
    """
    Retorna uma lista com os classificadores que serão avaliados.
    """
    classifiers = []
    classifiers.append(
                      ('knn',                              # nome do classificador
                        KNeighborsClassifier(),            # instancia do classificador
                        {'n_neighbors' : range(1, 33, 2)}  # hiperparametros
                      )
    )
    
    classifiers.append(
                      ('lr',                               # nome do classificador
                        LogisticRegression(max_iter=1000), # instancia do classificador
                        {'penalty' : ['l2'], 'C' : [100, 10, 1, 0.1, 0.01]}  # hiperparametros
                      )
    )
    
    classifiers.append(
                      ('svm',
                        SVC(max_iter=1000),
                        {'kernel' : ['linear', 'rbf'], 
                         'C' : [2 ** (-5), 1, 2 ** 5, 2 ** 10],
                         'gamma' : [2 ** (-15), 2 ** (-10), 2 ** (-5), 1, 2 ** 5]
                        }
                      )
    )
    
    classifiers.append(
                      ('dt',
                        DecisionTreeClassifier(),
                        {'max_depth' : [2, 4, 6, 8, 10, 12]}
                      )  
    )
    
    classifiers.append(
                      ('rf',
                        RandomForestClassifier(),
                        {'n_estimators' : [10, 50, 100]}
                      ) 
    )
    
    classifiers.append(
                      ('GaussianNB',
                        GaussianNB(),
                        {}
                      ) 
    )
    
    return classifiers
```

```python
classifiers = build_classifiers()
results = []
names = []

for name, model , parameters in classifiers:
    print('\n>> classifier name: {}'.format(name))
    gs = GridSearchCV(estimator=model, param_grid=parameters, refit=True, verbose=0)
    kf = KFold(n_splits=10, shuffle=True, random_state=RANDOM_STATE)
    cv = cross_val_score(gs, X_train, y_train, cv=kf, scoring='accuracy')
    print('Mean: {}, Standart Deviation: {}'.format(cv.mean(), cv.std()))
    #gs.best_params_
    #print(gs.best_score_)
    results.append(cv)
    names.append(name)
    # TODO: guardar a estatistica de teste para plotar um grafico de barras.
```

```python
fig, ax = plt.subplots()
fig.suptitle('Desempenho dos algoritmos em treinamento')
plt.boxplot(results)
ax.set_xticklabels(names)
ax.set_xlabel('Classificadores')
ax.set_ylabel('Acuracia')
plt.show()
```

# Passo 6. Teste do Modelo

```python
classifiers = build_classifiers()
results_testers = []
names = []

for name, model , parameters in classifiers:
    print('\n>> classifier name: {}'.format(name))
    gs = GridSearchCV(estimator=model, param_grid=parameters, refit=True, verbose=0)
    gs.fit(X_train, y_train)
    y_pred = gs.predict(X_test)
    acuracia_teste = gs.score(X_test, y_test)
    print('Acurácia do modelo: {}'.format(gs.score(X_test, y_test)))
    print(confusion_matrix(y_test, y_pred))
    print(classification_report(y_test,y_pred))
    
    # gs.best_params_
    # print(gs.best_score_)
    results_testers.append(acuracia_teste)
    names.append(name)
    # TODO: guardar a estatistica de teste para plotar um grafico de barras.
```

```python
fig, ax = plt.subplots()
fig.suptitle('Desempenho dos algoritmos em treinamento')
sns.barplot(names,results_testers)

ax.set_xlabel('Classificadores')
ax.set_ylabel('Acuracia')
```
